```python
import warnings
warnings.filterwarnings('ignore')

import ipywidgets as widgets
from IPython.display import display, clear_output
import math
from ipywidgets import interact, interactive
```

```python
#!jupyter nbextension enable - py widgetsnbextension - sys - prefix
#!jupyter serverextension enable voila - sys-prefix
```

```python
file = open("kinematics.png","rb")
image = file.read()

image_headline = widgets.Image(
                value=image,
                format='png',
                width='300')
```

```python
units = widgets.ToggleButtons(options=['US Customary','Metric'])
problem_choice = widgets.ToggleButtons(options=[('Velocity, Angle',1), ('Velocity Components',2)])
angle = widgets.IntSlider(min = 0, max = 90, step = 1)
velocity = widgets.Text(placeholder = 'Enter a launch velocity')
height = widgets.Text(placeholder = 'Enter a launch height')
x_velocity = widgets.Text(placeholder = 'Enter an x velocity')
y_velocity = widgets.Text(placeholder = 'Enter a y velocity')
```

```python
button_launch = widgets.Button(
                description = 'Calculate launch',
                tooltip = 'Launch')
```

```python
output = widgets.Output()
```

```python
def on_launch_clicked(event):
    global problem_choice
    with output:
        clear_output()
        x_distance, max_height, duration, display_units = calculate()
        
        print('3...2...1...Launch!')
        print(f"The projectile flew {x_distance} {display_units}")
        print(f"The max height was {max_height} {display_units}")
        print(f"The projectile was in the air for {duration} seconds")
button_launch.on_click(on_launch_clicked)
vbox_result = widgets.VBox([button_launch,output])
```

```python
def choose_problem(Kinematics):
    global choice
    if Kinematics == 2:
        choice = 2
        hbox_x = widgets.HBox([text_x_velocity, x_velocity,text_units])
        hbox_y = widgets.HBox([text_y_velocity,y_velocity,text_units])
        vbox_choice = widgets.VBox([hbox_x,hbox_y])
    else:
        choice = 1
        hbox_angle = widgets.HBox([text_angle, angle, text_degrees])
        hbox_velocity = widgets.HBox([text_velocity, velocity, text_units])
        vbox_choice = widgets.VBox([hbox_angle,hbox_velocity])
    return vbox_choice
```

```python
text_x_velocity = widgets.HTML(value="<h4>Enter X velocity</h4>")
text_units = widgets.HTML(value="<h4>(ft/s, m/s)</h4>")
text_y_velocity = widgets.HTML(value="<h4>Enter Y velocity</h4>")
text_angle = widgets.HTML(value="<h4>Enter a launch angle</h4>")
text_degrees = widgets.HTML(value="<h4>degrees</h4>")
text_velocity = widgets.HTML(value="<h4>Enter a launch velocity</h4>")
text_height = widgets.HTML(value="<h4>Enter a launch height</h4>")
text_height_units = widgets.HTML(value="<h4>(ft, m)</h4>")
text_Title = widgets.HTML(value="<h1>Calculate Launch!</h1>")
```

```python
def calculate():
    #Take inputs from widgets and return the calculations for the projectile motion
    #return x_distance, max_height, duration, units
    acc_us = -32.02 #ft/s^2
    acc_m = -9.8 #m/s^2
    if units.value == 'Metric':
        display_units = 'meters'
        acc = acc_m #set acceleration to use metric units
    else:
        display_units = 'feet'
        acc = acc_us #set acceleration to US Customary units
    #add try/except for all inputs
    if choice == 1:
        try:
            vel_x = float(velocity.value) * math.cos(math.radians(int(angle.value)))
            vel_y = float(velocity.value) * math.sin(math.radians(int(angle.value)))
        except:
            vel_x = 0
            vel_y = 0
    else:
        try:
            vel_x = float(x_velocity.value)
            vel_y = float(y_velocity.value)
        except:
            vel_x = 0
            vel_y = 0
    time_to_peak = -vel_y/acc
    try:
        float(height.value)
        launch_height = height.value
    except:
        launch_height = 0
    max_height = launch_height + (vel_y)*time_to_peak/2
    time_to_ground = math.sqrt(-max_height * 2/acc)
    duration = time_to_peak + time_to_ground
    x_distance = vel_x * duration
    return round(x_distance,2), round(max_height,2), round(duration,2), display_units
```

```python
hbox_height = widgets.HBox([text_height, height,text_height_units])
vbox_text = widgets.VBox([hbox_height, units, vbox_result ])
vbox_image = widgets.VBox([text_Title,image_headline])
display(vbox_image)
problem_choice = interact(choose_problem, Kinematics = widgets.ToggleButtons(options=[('Velocity, Angle',1), ('Velocity Components',2)]))
display(vbox_text)
```
